```python
#Librerías
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Leer y mostrar archivo
df = pd.read_excel('salas.xlsx', header=1)

filas, columnas = df.shape
print(f"Filas: {filas}\nColumnas: {columnas}")
df.head()
```

```python
#Eliminamos columnas inecesarias
df = df.drop(['num_enchufes','mts2','foto'], axis= 1)
df.head()
```

```python
cps = df['Campus'].unique()
print(cps)
#XD no sirve de nd
```

```python
df = df.drop('Campus', axis= 1)

print(df.shape)
df.head()
```

```python
# Primer piso: 
# Opcional: Cargar el excel desde 0 para evitar problemas con las casillas previas
# df = pd.read_excel('salas.xlsx', header=1)


# 1. Definicion de la regex para el primer piso del edificio F
# Como algunas salas empiezan con el prefijo “CORE” y otras no usamos la regex:
#    - ^                                : inicio de la celda
#    - (?:CORE\s*)?                    : opcionalmente “CORE ” (con o sin espacio)
#    - (10[1-9]|11[0-4])               : un número entre 101 y 114
#    - -F$                             : y que termine en “-F”
regex_primer_piso = r'^(?:CORE\s*)?(10[1-9]|11[0-4])-F$'

# 2. Aplicamos el filtro sobre la columna “Sala”
mask_primer_piso = df['Sala'].astype(str).str.match(regex_primer_piso)

# 3. Nos quedamos solo con esas filas
salas_primer_piso = df[mask_primer_piso].copy()

# 4. Verificación rápida
print(f"Filas totales:         {len(df)}")
print(f"Filas primer piso:     {len(salas_primer_piso)}")
display(salas_primer_piso)
```

```python
#Demostracion segundo piso
regex_segundo_piso = r'^(?:CORE\s*)?(20[1-9]|21[0-4])-F$'

mask_segundo_piso = df['Sala'].astype(str).str.match(regex_segundo_piso)

salas_segundo_piso = df[mask_segundo_piso].copy()

print(f"Filas totales:         {len(df)}")
print(f"Filas primer piso:     {len(salas_segundo_piso)}")
display(salas_segundo_piso)
```
